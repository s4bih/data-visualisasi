```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Z = 5
Y = 10.5

plt.plot(Z,Y, marker = 'o', markersize=5,color="r")
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]

plt.plot(x,y,'*--r')
plt.show()
```

```python
x = 1,2,3,4,5,6,7,8
y = 5,2,4,2,1,4,5,2

plt.plot(x,y,'*--r',lw=1)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```

```python
x = 1,2,3,4,5,6,7,8
y = 5,2,4,2,1,4,5,2
plt.figure(figsize=(10,10))

plt.plot(x,y,'*--r',lw=1)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```

##scatter plot


```python
plt.figure(figsize=(10,10))
x = 1,2,3,4,5,6,7,8
y = 5,2,4,2,1,4,5,2

colors = np.array(['r','g','b','c','m','y','k','w'])

plt.scatter(x,y, color = colors)
plt.title("Scatter plot")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.show()
```

```python
plt.figure(figsize=(10,10))
x = 1,2,3,4,5,6,7,8
y = 5,2,4,2,1,4,5,2

colors = np.array([20,30,40,50,60,70,80,90])

plt.scatter(x,y, c = colors)

plt.title("Scatter plot")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.show()
```

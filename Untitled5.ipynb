```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
url = 'https://drive.google.com/uc?id=1NI1zHuMTh6C9LY20KcM2DHBCXyHODbxH'
```

```python
vg_game = pd.read_csv(url)
vg_game
```

```python
vg_game['Global_Sales'].sort_values(ascending=False)
```

```python
vg_game.info()
```

```python
vg_game.describe()
```

```python
top_10_sales = vg_game[vg_game['Global_Sales'] > 10]
top_10_sales
```

```python
top_10_sales = vg_game[vg_game['Global_Sales'] > 28.31]
plt.bar(top_10_sales['Genre'], top_10_sales['Global_Sales'])
plt.xticks(rotation=90)
plt.xlabel('Genre')
plt.ylabel('Global Sales')
plt.title('Top 10 Games by Global Sales')
plt.show()
```

```python
vg_game.isnull().sum()
```

```python
vg_game.reset_index(drop=True, inplace=True)
```

```python
vg_game
```

```python
vg_game.dropna(inplace=True)
```

```python
vg_game[['NA_Sales',	'EU_Sales',	'JP_Sales',	'Other_Sales',	'Global_Sales']].dtypes
```

```python
vg_game['NA_Sales'] = pd.to_numeric(vg_game['NA_Sales'])
vg_game['EU_Sales'] = pd.to_numeric(vg_game['EU_Sales'])
vg_game['JP_Sales'] = pd.to_numeric(vg_game['JP_Sales'])
vg_game['Other_Sales'] = pd.to_numeric(vg_game['Other_Sales'])
vg_game['Global_Sales'] = pd.to_numeric(vg_game['Global_Sales'])
```

```python

vg_game['total_sales'] = vg_game['NA_Sales'] + vg_game['EU_Sales'] + vg_game['JP_Sales'] + vg_game['Other_Sales'] + vg_game['Global_Sales']
vg_game
```

```python
vg_game['total_sales'].sort_values(ascending=False)
```

```python
top_10_games = vg_game[vg_game['total_sales'] > 56]
plt.bar(top_10_games['Name'], top_10_games['total_sales'])
plt.xticks(rotation=90)
plt.xlabel('Name')
plt.ylabel('Total Sales')
top_10_games
```

```python
sns.barplot(x='Name', y='total_sales', data=top_10_games)
plt.xticks(rotation=90)
plt.xlabel('Name')
plt.ylabel('Total Sales')
plt.title('Top 10 Games by Total Sales')
```

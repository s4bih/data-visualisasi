```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data = 'https://drive.google.com/uc?id=1mVBXALH-5M83bMzL6I9YeE7-_fZ_P-OW'
```

```python
co2 = pd.read_csv(data)
co2
```

```python
co2.info()
```

```python
co2.describe()
```

```python
co2.isnull().sum()
```

```python
co2.dropna(inplace=True)
```

```python
co2.isnull().sum()
```

```python
co2.reset_index(drop=True, inplace=True)
co2
```

```python
co2.columns
```

```python
co2['Indicator Name'].value_counts()
```

```python
co2['2019'].equals(co2['2019.1'])
```

```python
co2.columns
```

```python
co2.drop(['2019.1'], axis=1, inplace=True)
co2
```

```python
co2['total_emissions'] = co2[['1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999','2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008','2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017','2018', '2019']].sum(axis=1)
#The .sum(axis=1) calculates the sum across the specified columns (1991-2019) for each row, resulting in a single value that can be assigned to the 'total_emissions' column.
```

```python
co2.head(10)
```

```python
region_total = co2.groupby('Region')['total_emissions'].sum().reset_index()
region_total
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x = 'Region', y = 'total_emissions', data = region_total)
plt.title('Total Emissions by Region')
plt.xlabel('Region')
plt.ylabel('Total Emissions')

plt.xticks(rotation=90)
```

```python
value = co2.sort_values(by='total_emissions', ascending=False).head(10)
value
```

```python
lab = value['Country Name']
val = value['total_emissions']
plt.figure(figsize=(20, 8))
plt.pie(val,labels=lab,autopct='%1.1f%%')
plt.title('Total Emissions by Country')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://drive.google.com/uc?id=1Ld-ez8xqx-ntpE7JgBZpSsECrrAOkRSR'
titanic = pd.read_csv(url)
```

```python
titanic.dropna(inplace=True)
titanic.reset_index(drop=True, inplace=True)
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
survived_data = titanic['Survived'].value_counts()
survived_data
```

```python
plt.figure(figsize = (10,8))
plt.pie(survived_data.values, labels=survived_data.index, autopct='%1.1f%%')
plt.title('Survived')
plt.show()
```

```python
sns.boxplot(data=titanic, x='Pclass', y='Age')
```

```python
plt.figure(figsize=(10,8))
sns.countplot(data=titanic, x='Pclass',hue = 'Survived')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flights_url = 'https://drive.google.com/uc?id=13XxxA7DMXH64KtdHeqWH4W6_lOVrcmOV'
airports_url = 'https://drive.google.com/uc?id=1LhmE6VUZoArDDk4bYn8AEU3dnqomCGqR'
flight  = pd.read_csv(flights_url)
airport = pd.read_csv(airports_url)
flight
```

```python
#SCATTER PLOt
plt.figure(figsize=(10,5))
sns.scatterplot( data = flight, x = 'ArrDelay', y = 'DepDelay')
plt.show()
```

```python
airplot_flight = flight.merge(airport, right_on = 'airport_id', left_on = 'OriginAirportID',how = 'inner')
airplot_flight
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data = airplot_flight, x = 'Carrier', y = 'DepDelay')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data = airplot_flight, x = 'state')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
url = 'https://drive.google.com/uc?id=1fPXIKVsQyNFYRl-pbPms07N1mQz16mWa'
```

```python
ytb = pd.read_csv(url)
```

```python
ytb.info()
```

```python
ytb.describe()
```

```python
ytb.isnull().sum()
```

```python
ytb.dropna(inplace=True)
ytb
```

```python
ytb.reset_index(drop=True, inplace=True)
ytb
```

```python
ytb.isnull().sum()
```

```python
value = ytb.sort_values(by='Followers', ascending=False).head(10)
value
```

```python
plt.figure(figsize = (10,6))
plt.bar(value['Channel Name'], value['Followers'])
plt.xticks(rotation=90)
plt.xlabel('Channel Name')
plt.ylabel('Followers')
```

```python
plt.figure(figsize = (10,10))
plt.bar(value['Username'], value['Followers'])
plt.xticks(rotation=45)
plt.xlabel('Username')
plt.ylabel('Followers')
```

```python
lab = value['Username']
size = value['Posts']
plt.figure(figsize = (10,10))
plt.pie(size, labels=lab, autopct='%1.1f%%')
plt.title('video/image post')
plt.show()
```

```python
sorted_data = ytb.sort_values(by='Followers', ascending=False)


top_10_us  = sorted_data[sorted_data['Country'] == ''].head(10)
top_10_us
```

```python
plt.figure(figsize = (10,10))


plt.bar(value['Username'], value['Followers'],label = 'global')
plt.bar(top_10_us['Username'], top_10_us['Followers'],label = 'BR')
plt.legend()
plt.xticks(rotation=45)
plt.title('global vs brazil')
plt.xlabel('Username')
plt.ylabel('Followers')
```

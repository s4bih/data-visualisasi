```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
url  = 'https://drive.google.com/uc?id=1fPXIKVsQyNFYRl-pbPms07N1mQz16mWa'
```

```python
insta_data  = pd.read_csv(url)
insta_data
```

```python
insta_data.info()
```

```python
insta_data.describe()
```

```python
value = insta_data['Followers'].sort_values(ascending=False)[:10]
value
```

```python
top_10 = insta_data[insta_data['Followers']>=value[9]]
top_10
```

```python
insta_data
```

```python
plt.figure(figsize= (20,8))
plt.pie(top_10['Followers'], labels=top_10['Username'], autopct='%1.1f%%')
plt.show()
```

```python
sorted_data = insta_data.sort_values(by='Followers', ascending=False)

top_10_overall = sorted_data.head(10)
top_10_indo  = sorted_data[sorted_data['Country'] == 'ID'].head(10)
top_10_indo
```

```python
plt.figure(figsize=(10,6))
plt.xticks(rotation=90)
plt.legend()
plt.title('Top 10 Instagram Users by Followers')
plt.xlabel('Username')
plt.ylabel('Followers')
plt.bar(top_10_overall['Username'], top_10_overall['Followers'], label='Overall')
plt.bar(top_10_indo['Username'], top_10_indo['Followers'], label='Indonesia')
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
plt.legend()

plt.xlabel('Username')
plt.ylabel('views')
plt.title('Top 10 Instagram Users by Views')
sns.barplot(x='Username', y='Views Avg.', data=top_10_overall, label='Overall')
sns.barplot(x='Username', y='Views Avg.', data=top_10_indo, label='Indonesia')
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = list(range(1,6))
print(x)
```

# **histogram**

```python
marks = np.random.randint(0,100,150)
print(marks)
```

```python
plt.xlabel('mark')
plt.ylabel('count')
plt.title('histogram')
plt.hist(marks,bins=25,edgecolor='black')
plt.grid(True)

```

# **pie chart**

```python
labels = ['phyton','c++','ruby','java']
sizes = [215,130,245,210]
color = ['yellowgreen','skyblue','lightcoral','gold']

plt.pie(sizes,labels=labels,autopct='%.2f%%',explode=(0.1,0,0,0)
,shadow=True,startangle=90,colors = color)
plt.axis('equal')
plt.legend()
plt.title('pie chart')
plt.show()
```

# **compare plot**

```python
x = np.linspace(0,10,100)
x
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1,label='sin(x)')
plt.plot(x,y2,label='cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('compare plot')
plt.legend()
plt.show()
```

# **subplot**

```python
plt.figure(figsize=(10,5))
plt.suptitle('2 subplot')

sub = ['math','science','lang','bio']
m1 = [80,70,23,56]
m2 = [12,34,54,56]

plt.subplot(1,2,1)
plt.bar(sub,m1)
plt.title('bar1')

plt.subplot(1,2,2)
plt.bar(sub,m2)
plt.title('bar2')

```

```python
activites = ['eat','sleep','work','play','other']
time_spent = [2,7,2,1,12]
colors = ['yellowgreen','skyblue','lightcoral','gold','red']


#create a figure with subplot
fig,axs =  plt.subplots(ncols=2,nrows=2,figsize=(10,10))
#plot on etch subplot
axs[0,0].bar(activites,time_spent,color=colors)
axs[0,0].set_title('subplot1')

axs[0,1].plot(activites,time_spent)
axs[0,1].set_title('subplot2')

axs[1,0].scatter(activites,time_spent,color=colors,marker='*')
axs[1,0].set_title('subplot3')

axs[1,1].pie(time_spent,labels=activites,autopct='%.2f%%',colors = colors)
axs[1,1].set_title('subplot4')
```

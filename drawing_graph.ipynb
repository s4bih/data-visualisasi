```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = list(range(1,6))
print(x)
```

```python
y = list(range(1,10,2))
print(y)
```

```python
plt.figure(figsize=(10,5))
plt.bar(x,y)

#add label
plt.xlabel('x')
plt.ylabel('y')
plt.title('Bar Chart')
plt.show()
```

```python
activitis = ['eat','sleep','work','play','other']
time_spent = [2,7,2,1,12]

plt.figure(figsize=(10,5))
plt.bar(activitis,time_spent)

#add label
plt.xlabel('activity')
plt.ylabel('time spent')
plt.title('activities Chart')
plt.show()
```

```python
activitis = ['eat','sleep','work','play','other']
time_spent = [2,7,2,1,12]

plt.figure(figsize=(10,5))
plt.bar(activitis,time_spent,color='r', width =0.2)

#add label
plt.xlabel('activity')
plt.ylabel('time spent')
plt.title('activities Chart')
plt.show()
```

```python
bulan = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
pendapatan = [15000,18000,20500,22000,19500,25000,28000,24500,23000,26000,30000,32500]

plt.figure(figsize=(10,5))
plt.bar(bulan,pendapatan,width = 0.5)

#add label
plt.xlabel('bulan')
plt.ylabel('pendapatan(usd)')
plt.title('pendapatan Chart')
plt.xticks(range(len(bulan)),bulan,rotation = 90)
plt.show()
```

```python

plt.figure(figsize=(10,5))
plt.barh(activitis,time_spent)

#add label
plt.xlabel('activity')
plt.ylabel('time spent')
plt.title('activities Chart')
plt.show()
```

# **histogram**

```python
marks = np.random.randint(0,100,150)
print(marks)
```

```python
plt.xlabel('mark')
plt.ylabel('count')
plt.title('histogram')
plt.hist(marks,bins=25,edgecolor='black')
plt.grid(True)

```

# **pie chart**

```python
labels = ['phyton','c++','ruby','java']
sizes = [215,130,245,210]
color = ['yellowgreen','skyblue','lightcoral','gold']

plt.pie(sizes,labels=labels,autopct='%.2f%%',explode=(0.1,0,0,0)
,shadow=True,startangle=90,colors = color)
plt.axis('equal')
plt.legend()
plt.title('pie chart')
plt.show()
```

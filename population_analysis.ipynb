```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://drive.google.com/uc?id=1id___NYaSi1FYR8N_12N-hCFtB2pJJs7'
population_data = pd.read_csv(url)
```

```python
population_data
```

```python
population_data.info()
```

```python
population_data.describe()
```

```python
top_10 = population_data.sort_values(by ='2022 Population',ascending=False ).head(10)
top_10
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data = top_10,x = 'Country/Territory',y = '2022 Population',marker='o',hue = 'Continent')

plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Top 10 Countries with Highest Population')
plt.xticks(rotation = 90)
plt.show()
```

```python
population = top_10.iloc[1,5:12]
population
```

```python
plt.figure(figsize=(16,6))
plt.bar(population.index,population)
plt.title('Top 10 Countries with Highest Population')
plt.xlabel('Population')
plt.ylabel('Country')
```
